```python
import pandas as pd

data = pd.read_csv('Supplement_Sales_Weekly_Expanded.csv')

data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
```

```python
data = data.drop_duplicates()
```

```python
print(data.isnull().sum())
```

```python
print(data['Category'].unique())
```

```python
print(data['Location'].unique())
```

```python
print(data[(data['Price'] <= 0) | (data['Units Sold'] < 0) | (data['Discount'] > 1)])
```

```python
data['Calculated Revenue'] = data['Units Sold'] * data['Price'] * (1 - data['Discount'])
discrepancies = data[data['Revenue'] != data['Calculated Revenue']]
print(discrepancies)
```

```python
data['Revenue'] = data['Calculated Revenue']
```

```python
data.to_csv('Supplement_Sales_Cleaned.csv', index=False)

from google.colab import files
files.download('Supplement_Sales_Cleaned.csv')
```
